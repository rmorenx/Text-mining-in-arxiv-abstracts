```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import torch
```

```python
model_name = 'sentence-transformers/allenai-specter'
INPUT_PATH= '/content/drive/MyDrive/arxiv/cleaned.csv'
```

```python
output_embeddings = "arxiv_embeddings.npy"  #matrix x
output_labels = "arxiv_labels.csv"          #etiqueta y

print("cargando dataset limpio....")
df = pd.read_csv(INPUT_PATH)

# Verificación de dispositivo
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"--- USANDO DISPOSITIVO: {device.upper()} ---")
if device == 'cpu':
    print("Sin GPU")

#cargar modelo
print(f"Cargando modelo {model_name}...")
model = SentenceTransformer(model_name, device=device)

#generar embeddings
print(f"iniciando vectorizacion de {len(df)} documentos...")

#sentencetransformer
embeddings = model.encode(
    df['cleaned_text'].tolist(),
    batch_size=128,
    show_progress_bar=True,
    convert_to_numpy=True
)

#resultados
print("guardar archivos en disco...")
np.save(output_embeddings, embeddings)

# Guardamos las etiquetas alineadas
df[['macro_cat', 'primary_cat']].to_csv(output_labels, index=False)

print("¡PROCESO TERMINADO!")
print(f"Matriz guardada: {embeddings.shape}")
```
